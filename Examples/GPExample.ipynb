```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from PrefInt.utility.gpranking import GPranking
from PrefInt.evaluate import ndcg
```

```python
# loaddata
loaddata = np.array(pd.read_csv('data/moldata/expmol.csv'))
expdata = loaddata[:,1:]
loaddata = np.array(pd.read_csv('data/moldata/simmol.csv'))
simdata = loaddata[:,1:]
#split train-test data
train,temp = train_test_split(expdata,test_size = 0.2, random_state = 1)
test = temp[:,:-1]
```

```python
#train: experimental data with known property values
#simdata: another dataset with imcompatible known property values
#test: experimental data need to be predicted with ranking
newtest,mu,sig2 = GPranking(train,simdata,test,musig=1)
print newtest
```

```python
gain = ndcg(mu,temp) #evaluating the ranking performance if the true property values of test data are known
print gain #Normalized Discounted Cumulative Gain
```
