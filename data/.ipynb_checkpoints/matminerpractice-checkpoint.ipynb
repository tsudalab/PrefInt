```python
import pandas as pd
import numpy as np

from matminer.featurizers.conversions import StrToComposition
from matminer.featurizers.composition import ElementProperty
```

```python
gwdata = pd.read_csv('../data/o/O_PBE.csv')
```

```python
gwdata.columns
```

```python
gwdata = gwdata.drop(['Unnamed: 0','id', 'ICSD SG', '&Delta;Ecbm',
       '&Delta;Evbm', '&epsilon;<sub>e</sub>(TDDFT)',
       'E<sub>g,d</sub> (eV)', 'm<sub>e</sub><sup>*</sup>/m<sub>0</sub>',
       'm<sub>h</sub><sup>*</sup>/m<sub>0</sub>', 'numSpin', 'standards',
       'keywords', 'parents'],axis=1)
```

```python
gwdata = StrToComposition().featurize_dataframe(gwdata, "sorted formula")
gwdata.head()
```

```python
gwdata = gwdata.drop_duplicates('composition')
```

```python
gwdata
```

```python
ep_feat = ElementProperty.from_preset(preset_name="magpie")
gwdata = ep_feat.featurize_dataframe(gwdata, col_id="composition")  # input the "composition" column to the featurizer
```

```python
gwdata
```

```python
gwdata = gwdata.drop(['final SG','composition','sorted formula'],axis=1)
```

```python
gwdata = gwdata.drop(['Unnamed: 0'],axis=1)
```

```python
c = gwdata.pop('E<sub>g</sub> (eV)')
```
